```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# layers = [4,8,16]
layers = [4,8,12]

models = ['BNT','BolT','Graphormer','NAGphormer','$NeuroPath$']
```

```python
layern = []
mn = []
offset = 0.5
for i in range(len(models)):
    layern.extend([l+i*offset for l in  layers])
    mn.extend([models[i] for _ in range(len(layers))])


data = {
    'F1 score': f1,
    'layer #': layern,
    'Model': mn,
}
data = pd.DataFrame(data)
print(data)
plt.figure(figsize=(3,5))
ax = sns.lineplot(data=data, x='layer #', y='F1 score', hue='Model', marker='*', markersize=20)
plt.grid()
plt.setp(ax.get_legend().get_texts(), fontsize='14') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='15') # for legend title
plt.savefig(f'{datan}_lineplot_ablation_layer.svg')
```
